```python
import nltk
from nltk.corpus import stopwords ##imported stopwords module

# Download the stopwords dataset (if you haven't already)
nltk.download('stopwords') ## downloaded the stopwords dataset(stopwords are most commomn words in the dataset with no meaning)

# Define STOPWORDS as a set of English stopwords
STOPWORDS = set(stopwords.words('english'))
```

```python
def clean_text(text):
    # convert to lowercase
    text = text.lower()
    # remove special characters
    text = re.sub(r'[^0-9a-zA-Z]', ' ', text)
    # remove extra spaces
    text = re.sub(r'\s+', ' ', text)
    # remove stopwords
    text = " ".join(word for word in text.split() if word not in STOPWORDS)
    return text
```

```python
## Cleaning the text in the message column and storing it in the new column: clean_text in the dataframe.
```

```python
X = df['clean_text']
y = df['label']
```

```python
from sklearn.pipeline import Pipeline
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.metrics import classification_report
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer, TfidfTransformer

def classify(model, X, y):
    # train test split
    x_train, x_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42, shuffle=True, stratify=y)
    # model training
    pipeline_model = Pipeline([('vect', CountVectorizer()),  ##  converts text data into a term frequency matrix.
                              ('tfidf', TfidfTransformer()),  ## transforms the term frequency matrix into a TF-IDF representation
                              ('clf', model)])
    pipeline_model.fit(x_train, y_train)

    print('Accuracy:', pipeline_model.score(x_test, y_test)*100)

#     cv_score = cross_val_score(model, X, y, cv=5)
#     print("CV Score:", np.mean(cv_score)*100)
    y_pred = pipeline_model.predict(x_test)
    print(classification_report(y_test, y_pred))
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
classify(model, X, y)
```

```python
## Multinomial Naive Bayes classifier for text classification
```

```python
from sklearn.naive_bayes import MultinomialNB
model = MultinomialNB()
classify(model, X, y)
```

```python
## SVM model
```

```python
from sklearn.svm import SVC
model = SVC(C=3)
classify(model, X, y)
```

```python
## Random Forest Classifier
```

```python
from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier()
classify(model, X, y)
```
